```python
%matplotlib notebook
%config InlineBackend.figure_formats = {'svg',}
import pandas
import matplotlib.pyplot as plt
import numpy as np
import datetime
import matplotlib.dates as mdates

data=pandas.read_csv("query1.csv")
print(data)
```

```python
histD=[]
print(data[:1])
for x in data.itertuples():
    year= (getattr(x,"Year"))
    month= (getattr(x,"Month"))
    day= (getattr(x,"Day"))
    mydate = datetime.date(year,month, day)  #year, month, day   
    histD.append(mydate)
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
percentage = data['PercCancelled']
#print (histD)
legend = ['Percentage of cancelled flights per day']
plt.plot(histD, percentage, marker='|', linestyle='-', color='r', label='Square')
#plt.hist([1,2], color=['blue'])
plt.xlabel("Day")
plt.ylabel("Frequency")
plt.legend(legend)
plt.title('Percentage of cancelled flights  per day')
plt.savefig('test.png', bbox_inches='tight')
plt.show()
```
